```python
# Carpeta donde estÃ¡n los CSV
ruta = "../dataset_ecobici/"
archivos = glob.glob(os.path.join(ruta, "*.csv"))

df_total = pd.DataFrame()
total_filas = 0
contador_archivos = 0

print(f"ğŸ” Se encontraron {len(archivos)} archivos para procesar.\n")

for archivo in archivos:
    contador_archivos += 1
    nombre = os.path.basename(archivo)
    print(f"ğŸ“‚ Procesando archivo {contador_archivos}/{len(archivos)}: {nombre}")

    # Leer en bloques (chunks)
    contador_chunks = 0
    for chunk in pd.read_csv(archivo, chunksize=100_000):
        contador_chunks += 1
        filas_chunk = len(chunk)
        total_filas += filas_chunk

        # Concatenar incrementalmente
        df_total = pd.concat([df_total, chunk], ignore_index=True)

        # Calcular RAM usada
        uso_ram = psutil.virtual_memory().percent

        print(f"   ğŸ§© Chunk {contador_chunks}: {filas_chunk:,} filas | "
              f"Total: {total_filas:,} | RAM usada: {uso_ram:.2f}%")

        # Liberar memoria entre iteraciones
        gc.collect()

print("\nâœ… Lectura completada.")
print(f"ğŸ“Š Total de archivos: {contador_archivos}")
print(f"ğŸ“ˆ Total de filas leÃ­das: {total_filas:,}")
print(f"ğŸ§  RAM final usada: {psutil.virtual_memory().percent:.2f}%")
print(f"ğŸ“¦ TamaÃ±o del DataFrame en memoria: {df_total.memory_usage(deep=True).sum() / (1024**2):.2f} MB")
```

```python
import os, psutil, pandas as pd, glob, gc, time

def humanize_bytes(n):
    for unit in ["B","KB","MB","GB","TB"]:
        if n < 1024:
            return f"{n:,.2f} {unit}"
        n /= 1024
    return f"{n:,.2f} PB"

# Ajusta tu ruta
archivos = sorted(glob.glob("../dataset_ecobici/*.csv"))
proceso = psutil.Process(os.getpid())
total_filas = 0
chunks = []
t0 = time.time()

print(f"RAM inicial: {humanize_bytes(proceso.memory_info().rss)}")

for f_idx, archivo in enumerate(archivos, 1):
    for c_idx, chunk in enumerate(pd.read_csv(archivo, chunksize=100_000), 1):
        chunks.append(chunk)  # <- esto hace crecer la RAM a propÃ³sito
        total_filas += len(chunk)
        rss = proceso.memory_info().rss
        print(f"[Archivo {f_idx}/{len(archivos)} | Chunk {c_idx}] "
              f"Filas acumuladas: {total_filas:,} | RAM: {humanize_bytes(rss)}")

df_total = pd.concat(chunks, ignore_index=True)
print(f"RAM tras concat: {humanize_bytes(proceso.memory_info().rss)}")
print(f"Tiempo total: {time.time()-t0:,.1f}s | Filas: {len(df_total):,}")
```

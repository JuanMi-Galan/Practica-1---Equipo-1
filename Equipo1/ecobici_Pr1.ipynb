## Definicion de funciones

```python
# Defini la funcion para ver la completitud de los datos
def completitud_datos(df):
    return round(df.isnull().sum().sort_values(ascending=False) / df.shape[0], 4)
```

```python
def expresion_regular_hora(df):
    df['Hora_Retiro'] = df['Hora_Retiro'].astype(str).str.split('.').str[0]
    df['Hora_Arribo'] = df['Hora_Arribo'].astype(str).str.split('.').str[0]
    df['Hora_Retiro'] = (
        df['Hora_Retiro']
        .astype(str)
        .str.replace(r'^(\d{1,2})[:]{1,2}$', r'\1:00:00', regex=True)  # Corrige casos como 18:: o 9:
    )
    df['Hora_Arribo'] = (
        df['Hora_Arribo']
        .astype(str)
        .str.replace(r'^(\d{1,2})[:]{1,2}$', r'\1:00:00', regex=True)  # Corrige casos como 18:: o 9:
    )
```

```python
# Columnas para todos los datos
columnas_requeridas = [
    "Genero_Usuario", "Edad_Usuario", "Bici",
    "Ciclo_Estacion_Retiro", "Fecha_Retiro", "Hora_Retiro",
    "Ciclo_Estacion_Arribo", "Fecha_Arribo", "Hora_Arribo"
]
```

```python
dtypes_dict = {"Genero_Usuario": 'string', "Edad_Usuario": 'int', "Bici": 'string',
    "Ciclo_Estacion_Retiro": 'string', "Fecha_Retiro": 'string', "Hora_Retiro": 'string',
    "Ciclo_Estacion_Arribo": 'string', "Fecha_Arribo": 'string', "Hora_Arribo": 'string'}
```

```python
def rango_apertura_fechas(df):
    fecha_min = df['Fecha_Arribo'].min()
    fecha_max = df['Fecha_Arribo'].max()
    return fecha_min, fecha_max
```

```python
def formato_fechas_horas(df):
    # Unir fecha y hora
    df['FechaHora_Retiro'] = pd.to_datetime(
        df['Fecha_Retiro'].astype(str) + ' ' + df['Hora_Retiro'].astype(str),
        errors='coerce'
    )

    # Unir fecha y hora
    df['FechaHora_Arribo'] = pd.to_datetime(
        df['Fecha_Arribo'].astype(str) + ' ' + df['Hora_Arribo'].astype(str),
        errors='coerce'
    )
```

```python
# Calcular la duracion en minutos
def calcular_duracion(df):
    df['Duracion'] = (df['FechaHora_Arribo'] - df['FechaHora_Retiro']).dt.total_seconds() / 60
```

```python
def exportar_parquet_limpio(df, anio):
    df[["Genero_Usuario", "Edad_Usuario", "Bici",
    "Ciclo_Estacion_Retiro", "FechaHora_Retiro",
    "Ciclo_Estacion_Arribo", "FechaHora_Arribo", "Duracion"]].to_parquet(f"../ecobici_data_segmentados/ecobici_{anio}.parquet", index=False)
```
